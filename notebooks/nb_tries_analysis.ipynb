```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# open csv using pandas
lib_sample_results = "../results_llm/stats_llm_libraries_sample10_context.csv"
cedar_sample_results = "../results_llm/stats_llm_cedar_sample10_context.csv"
vmc_sample_results = (
    "../results_llm/stats_llm_DafnyVMC_sample10_error_n_feedback_6t.csv"
)

df_cedar = pd.read_csv(cedar_sample_results)
df_lib = pd.read_csv(lib_sample_results)
df_vmc = pd.read_csv(vmc_sample_results)

df = pd.concat([df_cedar, df_lib, df_vmc], ignore_index=True)
df.head()
```

```python
df["Try"] = df["Try"] + 1
df_correct = df[df["New Method Result"] == "Correct"]
df_correct.head()
```

```python
cumulative_counts = df_correct["Try"].value_counts().sort_index().cumsum()
cumulative_counts = cumulative_counts.reindex(range(1, 11)).ffill()
print(cumulative_counts)

cumulative_counts.plot(kind="line", marker="o", figsize=(12, 6))
plt.xlabel("Number of Tries")
plt.ylabel("Cumulative Number of Correct Results")
plt.grid(True)
plt.show()
```
